```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function g(t)
def gaussian(t, t_H):
    return np.exp(-(t/t_H)**2) / (np.sqrt(np.pi) * t_H)

dt = 0.001
t_H1 = 5
t_H2 = 15

# Define the time domain
t = np.arange(-50, 50, dt)

# Calculate the Gaussian functions
g1 = gaussian(t, t_H1)
g2 = gaussian(t, t_H2)

# Plot the Gaussian functions
plt.plot(t, g1, label="t_H = 5 sec")
plt.plot(t, g2, label="t_H = 15 sec")

# Add legend and labels
plt.legend()
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.title("Gaussian function g(t)")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function g(t)
def gaussian(t, t_H):
    return np.exp(-(t/t_H)**2) / (np.sqrt(np.pi) * t_H)

# Define the time domain
t = np.arange(-50, 50, 0.001)

# Define the half durations
t_H1 = 5
t_H2 = 15

# Calculate the Gaussian functions
g1 = gaussian(t, t_H1)
g2 = gaussian(t, t_H2)

# Shift g(t) to start from zero
g1_shift = np.roll(g1, len(g1)//2)
g2_shift = np.roll(g2, len(g2)//2)

# Calculate the sampling frequency and frequency axis
fs = 1/(t[1]-t[0])
f_axis = np.fft.fftfreq(len(t), 1/fs)
w_axis = 2*np.pi*f_axis

# Calculate the DFT of g(t)
G1 = np.fft.fftshift(np.fft.fft(g1_shift))
G2 = np.fft.fftshift(np.fft.fft(g2_shift))

# Calculate the analytical Fourier transform of g(t)
G1_analytical = np.exp(-(w_axis**2 * t_H1**2)/4)
G2_analytical = np.exp(-(w_axis**2 * t_H2**2)/4)

# Plot the DFT and analytical Fourier transform
plt.plot(w_axis, np.abs(G1), label="t_H = 5 sec (DFT)")
plt.plot(w_axis, np.abs(G1_analytical), label="t_H = 5 sec (Analytical)")
plt.plot(w_axis, np.abs(G2), label="t_H = 15 sec (DFT)")
plt.plot(w_axis, np.abs(G2_analytical), label="t_H = 15 sec (Analytical)")

# Add legend and labels
plt.legend()
plt.xlabel("Frequency (rad/sec)")
plt.ylabel("Amplitude")
plt.title("DFT and Analytical Fourier Transform of Gaussian function g(t)")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function
def gaussian(t, th):
    return np.exp(-(t/th)**2) / (np.sqrt(np.pi) * th)

# Define the time axis
t = np.arange(-50, 50, 0.001)

# Define the half durations
th1 = 5
th2 = 15

# Compute the Gaussian functions
g1 = gaussian(t, th1)
g2 = gaussian(t, th2)

# Shift the Gaussian functions to start at time 0
g1_shift = np.roll(g1, len(g1)//2)
g2_shift = np.roll(g2, len(g2)//2)

# Compute the DFT of the shifted Gaussian functions
G1 = np.fft.fftshift(np.fft.fft(g1_shift)) * \
        np.exp(-1j * 2 * np.pi * t[0] * np.fft.fftfreq(len(t), d=0.001))
G2 = np.fft.fftshift(np.fft.fft(g2_shift)) * \
        np.exp(-1j * 2 * np.pi * t[0] * np.fft.fftfreq(len(t), d=0.001))

# Compute the theoretical Fourier transforms
f_axis = np.fft.fftfreq(len(t), d=0.001)
w_axis = 2 * np.pi * f_axis
G1_theory = np.exp(-w_axis**2 * th1**2 / 4)
G2_theory = np.exp(-w_axis**2 * th2**2 / 4)

# Plot the Gaussian functions and their Fourier transforms
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.plot(t, g1_shift, label='tH = 5')
plt.plot(t, g2_shift, label='tH = 15')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('g(t)')

plt.subplot(2, 2, 2)
plt.plot(w_axis, np.abs(G1), label='DFT (tH = 5)')
plt.plot(w_axis, np.abs(G2), label='DFT (tH = 15)')
plt.plot(w_axis, np.abs(G1_theory), '--', label='Theory (tH = 5)')
plt.plot(w_axis, np.abs(G2_theory), '--', label='Theory (tH = 15)')
plt.legend()
plt.xlabel('Frequency (rad/s)')
plt.ylabel('|G(w)|')

plt.subplot(2, 2, 4)
plt.plot(f_axis, np.abs(G1), label='DFT (tH = 5)')
plt.plot(f_axis, np.abs(G2), label='DFT (tH = 15)')
plt.plot(f_axis, np.abs(G1_theory), '--', label='Theory (tH = 5)')
plt.plot(f_axis, np.abs(G2_theory), '--', label='Theory (tH = 15)')
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('|G(f)|')

plt.tight_layout()
plt.show()

```
