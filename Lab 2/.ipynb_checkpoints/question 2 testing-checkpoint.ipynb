```python
# Question 2 - a)

import numpy as np
import matplotlib.pyplot as plt

# Define the time axis
t = np.arange(0, 10, 0.01)

# Define the boxcar window function
boxcar = np.zeros_like(t)
boxcar[(t >= 0) & (t <= 10)] = 1

# Define the Hann window function
hann = 0.5*(1 - np.cos(2*np.pi*t/10))
hann[t > 10] = 0

# Plot the window functions on the same graph
plt.plot(t, boxcar, label='Boxcar')
plt.plot(t, hann, label='Hann')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
# Question 2 - b) 

import numpy as np
import matplotlib.pyplot as plt

# Define the time axis
t = np.arange(0, 10, 0.01)

# Define the boxcar window function
boxcar = np.zeros_like(t)
boxcar[(t >= 0) & (t <= 10)] = 1

# Define the Hann window function
hann = 0.5*(1 - np.cos(2*np.pi*t/10))
hann[t > 10] = 0

# Calculate the Fourier transform of the boxcar window function
boxcar_fft = np.fft.fft(boxcar)
boxcar_fft_shift = np.fft.fftshift(boxcar_fft)
freq = np.fft.fftfreq(len(boxcar), 0.01)
freq_shift = np.fft.fftshift(freq)

# Calculate the Fourier transform of the Hann window function
hann_fft = np.fft.fft(hann)
hann_fft_shift = np.fft.fftshift(hann_fft)

# Plot the Fourier transforms of the window functions
plt.subplot(2,1,1)
plt.plot(freq_shift, np.abs(boxcar_fft_shift))
plt.title('Boxcar Window Function Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.ylim(0, 1100)
plt.xlim(-1,1)

plt.subplot(2,1,2)
plt.plot(freq_shift, np.abs(hann_fft_shift))
plt.title('Hann Window Function Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.ylim(0, 1100)
plt.xlim(-1,1)

plt.tight_layout()
plt.show()
```

### Questino 2 - part d)
Both window functions (boxcar and Hann) effectively truncate the continuous time signal to a finite length, resulting in a non-zero signal only within the interval [0, T]. The effect of this truncation on the frequency spectrum of the original signal can be observed through the Fourier transforms of the window functions.

In general, the Fourier transform of a truncated signal will result in a convolution of the original Fourier transform with the Fourier transform of the window function. This convolution results in a smoothing effect on the frequency spectrum, with the degree of smoothing determined by the shape of the window function.

For the boxcar window, we can see that its Fourier transform is a sinc function with oscillations around the zeros. The sinc function has a relatively sharp transition between the passband and stopband, resulting in relatively high side lobes. The side lobes can be seen in the Fourier transform of the boxcar window function, where the amplitude does not fully drop to zero at high frequencies. This means that the boxcar window will result in significant leakage and distortion in the frequency spectrum of the original signal.

On the other hand, the Hann window function has a smooth transition between the passband and stopband, resulting in lower side lobes compared to the boxcar window. This can be seen in the Fourier transform of the Hann window, where the amplitude drops to zero at higher frequencies more rapidly than the boxcar window. Therefore, the Hann window will result in less leakage and distortion in the frequency spectrum of the original signal compared to the boxcar window.

Overall, windowing a continuous time signal results in a trade-off between leakage and resolution in the frequency domain. The Hann window provides a good balance between these two factors and is often used in practice to minimize leakage while maintaining reasonable resolution.

### Question 3 - part e) 

The choice of window function for truncating a continuous time signal depends on the specific requirements of the application. Here are some advantages and disadvantages of boxcar and Hann window functions for truncation:

Boxcar window:

Advantages:
- The boxcar window function is simple and easy to implement.
- The boxcar window is rectangular in shape, which means that it provides a uniform weighting to all samples within the truncated interval. This can be useful when all samples within the truncated interval are equally important.

Disadvantages:
- The boxcar window has high side lobes, which can result in significant leakage and distortion in the frequency spectrum of the original signal.
- The rectangular shape of the boxcar window can lead to ringing artifacts in the time domain, which can introduce additional noise in the signal.


Hann window:

Advantages:
- The Hann window has a smooth transition between the passband and stopband, which results in lower side lobes compared to the boxcar window. This can minimize leakage and distortion in the frequency spectrum of the original signal.
- The Hann window is symmetric, which means that it preserves the phase information of the original signal.

Disadvantages:
- The Hann window is more complex than the boxcar window, which can make it more difficult to implement.
- The Hann window has a slower roll-off compared to other window functions such as the Hamming or Blackman windows. This can result in reduced frequency resolution.


Overall, the choice between boxcar and Hann window functions (or other window functions) depends on the specific requirements of the application. The Hann window is a good general-purpose window that provides a good balance between leakage and resolution. However, if high resolution or a specific spectral shape is required, other window functions may be more appropriate.
